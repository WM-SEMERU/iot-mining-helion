```python
'''
Calculating most common test files based on their presence on different app-integration test codes
+ Helper automation to open 10 random examples of each test files. 
'''
```

```python
import os
from os.path import join as pjoin
import random
```

```python
# repo="gpt_test_code_inspection"
repo=pjoin("integration_samples","core","tests","components")
test_files={}
test_files_dir={}
```

```python
for app in os.listdir(repo):
    if not os.path.isfile(pjoin(repo,app)):
        for file in os.listdir(pjoin(repo,app)):
            if file not in test_files:
                test_files[file]=1
                test_files_dir[file]=[pjoin(repo,app,file)]
            else:
                test_files[file]+=1
                test_files_dir[file].append(pjoin(repo,app,file))
```

```python
sorted_test_files=dict(sorted(test_files.items(), key=lambda item: item[1], reverse=True))
```

```python
sorted_test_files
```

```python
import subprocess
import keyboard
import pygetwindow as gw
import time

# Path to the Notepad++ executable
notepad_plus_plus_path = "C:\\Program Files\\Notepad++\\notepad++.exe"
for filename in sorted_test_files.keys():
    # List of file paths you want to open
    file_paths = random.sample(test_files_dir[filename],10)
    
    # Construct the command to open files in Notepad++
    command = [notepad_plus_plus_path]
    command.extend(file_paths)

    # Use subprocess to run the command
    process = subprocess.Popen(command)
    
    print(filename)
    print()
    # Wait for the 'q' key press
    print("Press 'q' to close Notepad++ and its tabs.")
    keyboard.wait('q')
    print()
    # Close all Notepad++ windows
    for window in gw.getWindowsWithTitle("Notepad++"):
        window.close()

    # Wait for the windows to close
    time.sleep(2)

    # Kill the Notepad++ process
    process.kill()
```
