```python
'''
Calculating most common test files based on their presence on different app-integration test codes
+ Helper automation to open 10 random examples of each test files. 
'''
```

```python
import os
from os.path import join as pjoin
from pathlib import Path
import random
```

```python
# repo="gpt_test_code_inspection"
repo_path = "/Users/daniel/code/wam/IoT/core/tests/components"
path = Path(repo_path)
test_files={}
test_files_dir={}
```

```python
for app in os.listdir(path):
    if not os.path.isfile(pjoin(path,app)):
        for file in os.listdir(pjoin(path,app)):
            if file not in test_files:
                test_files[file]=1
                test_files_dir[file]=[pjoin(path,app,file)]
            else:
                test_files[file]+=1
                test_files_dir[file].append(pjoin(path,app,file))
```

```python
sorted_test_files=dict(sorted(test_files.items(), key=lambda item: item[1], reverse=True))
```
